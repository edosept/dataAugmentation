```python
from torchvision import transforms
from PIL import Image
```

# Data Augmentation

augmentasi ini sedikit mengubah data asli, dengan tujuan:
- untuk lebih memperkaya data kita<br>
- meningkatkan generality (mengurangi overfit)<br>
- menambah data<br>

```python
img = Image.open('ImgTrain_1750.jpg') #from plant-pathology-2020-fgvc7
img
```

### Random Crop

```python
crop = transforms.RandomCrop(256)
crop(img)
```

### Random Flip

Flipping adalah transformasi sederhana yang membolak balikan gambar baik secara horizontal maupun vertical

```python
hflip = transforms.RandomHorizontalFlip()
hflip(img)
```

ada beberapa gambar yang ga boleh pakai RandomHorizontalFlip yaitu rambu jalan karena nanti makna nya berubah<br>
contoh lainnya X-Ray seperti jantung, paru paru masa gambarnya dibolak balik wkwkwk

```python
vflip = transforms.RandomVerticalFlip()
vflip(img)
```

image yang ga boleh RandomVerticalFlip --> landscape gedung, gunung.<br>
gedung dibalik gunung dibalik jadi langitnya dibawah kan ngaco jadinya wkwkwk

`jadi pada intinya itu kita pakai RandomFlip itu sesuai kebutuhan aja, kira kira perlu gga ya kita pake. jadi harus dipikirin`

### Random Rotation

```python
rotate = transforms.RandomRotation(10)
rotate(img)
```

RandomRotation --> tujuannya supaya gambar yang agak miring miring dikit mesinnya tetap belajar

### Random ResizedCrop

```python
resize = transforms.RandomResizedCrop(256)
resize(img)
```

### Random Affine

Random affine merupakan transformasi geser gambar secara bebas kekiri maupun kekanan

```python
translate = transforms.RandomAffine(0, translate=(0.3, 0.3))
translate(img)
```

```python
shear = transforms.RandomAffine(0, shear=10)
shear(img)
```

```python
affine = transforms.RandomAffine(5, translate=(0.1, 0.1), shear=10)
affine(img)
```

### Color Jitter

brightness nya, contrastnya kita atur untuk menghandle kemungkinan jika object tersebut ada dimalam hari misalnya menjadi agak gelap kan pasti objectnya, jadi walaupun malam hari object tersebut tetap terdeteksi.

```python
brightness = transforms.ColorJitter(brightness=0.1)
brightness(img)
```

```python
contrast = transforms.ColorJitter(contrast=0.1)
contrast(img)
```
